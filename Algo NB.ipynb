```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix,
                              f1_score, mean_squared_error, r2_score,
                              mean_absolute_error, roc_auc_score)
from xgboost import XGBRegressor
import lightgbm as lgb
from sklearn.model_selection import GroupShuffleSplit
# ======================
# 📥 Load Data
# ======================
df = pd.read_csv("/content/skycipher_dataset_20260512_125542.csv")

# ======================
# 🧹 Preprocessing — كل ما يلي يُطبق على كل البيانات (قبل السبليت)
# لأنه لا يعتمد على إحصاءات من البيانات
# ======================
df_clf = df.drop(["episode","bearing_deg","action_magnitude","target_dist",
                  "step", "reward","action_x", "action_y", "action_z", "target_dist", "rel_x", "rel_y", "rel_z", "row_id", "frame_id", "image_path", "full_img_path", "has_image", "label"], axis=1, errors="ignore")

# label mapping: تحويل نصوص لأرقام — لا data leakage
df_clf["episode_outcome"] = df_clf["episode_outcome"].map({
    "success": 1,
    "timeout": 0,
    "crash"  : 0,
    "stopped": 0
}).fillna(-1)

df_clf = df_clf.dropna()

# حذف أعمدة unique IDs — لا تعتمد على إحصاءات
for col in df_clf.columns:
    if df_clf[col].nunique() == len(df_clf):
        df_clf = df_clf.drop(col, axis=1)

# get_dummies: encoding — لا data leakage لأنه لا يعتمد على توزيع القيم
df_clf = pd.get_dummies(df_clf)

# ======================
# 🎯 Classification Split
# ======================
X_clf = df_clf.drop("episode_outcome", axis=1)
y_clf = df_clf["episode_outcome"]

groups = df.loc[X_clf.index, "episode"].values

gss = GroupShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=42
)

train_idx, test_idx = next(gss.split(X_clf, y_clf, groups))

X_train = X_clf.iloc[train_idx]
X_test  = X_clf.iloc[test_idx]

y_train = y_clf.iloc[train_idx]
y_test  = y_clf.iloc[test_idx]

print("✅ Group split applied using episode")

# ✅ StandardScaler يُطبق fit على Train فقط — هذا هو الصح
scaler_clf = StandardScaler()
X_train_s = scaler_clf.fit_transform(X_train)   # fit + transform على Train
X_test_s  = scaler_clf.transform(X_test)         # transform فقط على Test

# ======================
# 🤖 Classification Models
# ======================
models = {
    "Random Forest"      : RandomForestClassifier(n_estimators=100, random_state=42, class_weight="balanced"),
    "SVM"                : SVC(kernel="rbf", random_state=42, class_weight="balanced",
                               max_iter=5000, probability=True),
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight="balanced"),
    "Gradient Boosting"  : GradientBoostingClassifier(n_estimators=80, random_state=42)
}

results     = {}
preds       = {}
proba_preds = {}

SVM_SAMPLE = 20000
idx = np.random.RandomState(42).choice(len(X_train_s),
      size=min(SVM_SAMPLE, len(X_train_s)), replace=False)
X_train_svm = X_train_s[idx]
y_train_svm = y_train.iloc[idx]

# ======================
# 🏋️ Training + Evaluation
# ======================
print("=" * 55)
print("🔵 CLASSIFICATION — episode_outcome (0=Fail / 1=Success)")
print("=" * 55)

for name, model in models.items():
    if name == "SVM":
        model.fit(X_train_svm, y_train_svm)
        pred  = model.predict(X_test_s)
        proba = model.predict_proba(X_test_s)[:, 1]
        cv_scores = cross_val_score(model, X_train_svm, y_train_svm, cv=3, scoring="accuracy")
    elif name == "Logistic Regression":
        model.fit(X_train_s, y_train)
        pred  = model.predict(X_test_s)
        proba = model.predict_proba(X_test_s)[:, 1]
        cv_scores = cross_val_score(model, X_train_s, y_train, cv=5, scoring="accuracy")
    else:
        model.fit(X_train, y_train)
        pred  = model.predict(X_test)
        proba = model.predict_proba(X_test)[:, 1]
        cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")

    preds[name]       = pred
    proba_preds[name] = proba

    acc    = accuracy_score(y_test, pred)
    f1_mac = f1_score(y_test, pred, average="macro")
    f1_wei = f1_score(y_test, pred, average="weighted")
    mse    = mean_squared_error(y_test, pred)
    mae    = mean_absolute_error(y_test, pred)
    r2     = r2_score(y_test, pred)
    auc    = roc_auc_score(y_test, proba)

    results[name] = acc

    print(f"\n📌 {name}")
    print(f"  Accuracy          : {acc:.4f}")
    print(f"  F1 (macro)        : {f1_mac:.4f}")
    print(f"  F1 (weighted)     : {f1_wei:.4f}")
    print(f"  ROC-AUC           : {auc:.4f}")
    print(f"  MSE               : {mse:.4f}")
    print(f"  MAE               : {mae:.4f}")
    print(f"  R²                : {r2:.4f}")
    print(f"  Cross-Val Accuracy: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
    print(classification_report(y_test, pred))

# ======================
# 🔍 Feature Importance — Random Forest
# ======================
print("\n🔍 أهم 10 Features — Random Forest:")
feat_imp = pd.Series(models["Random Forest"].feature_importances_, index=X_train.columns)
print(feat_imp.sort_values(ascending=False).head(10))

plt.figure(figsize=(8, 5))
feat_imp.sort_values(ascending=False).head(10).plot(kind="bar", color="#4C72B0")
plt.title("Top 10 Feature Importances — Random Forest")
plt.ylabel("Importance")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# ======================
# 🏆 Classification Ranking
# ======================
print("\n🏆 Classification Ranking:")
for k, v in sorted(results.items(), key=lambda x: x[1], reverse=True):
    print(f"  {k}: {v:.4f}")

plt.figure(figsize=(7, 4))
plt.bar(results.keys(), results.values(), color="#4C72B0")
plt.title("Model Comparison — Classification Accuracy")
plt.xticks(rotation=20)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# ======================
# 🔥 Best Model Confusion Matrix
# ======================
best = max(results, key=results.get)
cm   = confusion_matrix(y_test, preds[best])
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix — Best Model: {best}")
plt.show()


# ============================================================
# 🟠 REGRESSION — التنبؤ بالـ Reward
# ============================================================
print("\n" + "=" * 55)
print("🟠 REGRESSION — Reward Prediction")
print("=" * 55)

df_reg = df.drop(["episode", "step",
                   "episode_outcome", "action_X", "action_Y", "action_Z"],
                  axis=1, errors="ignore")
df_reg = df_reg.dropna()

for col in df_reg.columns:
    if col != "reward" and df_reg[col].nunique() == len(df_reg):
        df_reg = df_reg.drop(col, axis=1)

X_reg = pd.get_dummies(df_reg.drop("reward", axis=1))
y_reg = df_reg["reward"]

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

def adjusted_r2(r2, n, k):
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)

n_test = len(y_test_r)
k_feat = X_test_r.shape[1]

# ✅ استبدلنا Random Forest Regressor بـ LightGBM — أسرع بكثير وأداء ممتاز
reg_models = {
    "XGBoost Regressor" : XGBRegressor(n_estimators=200, learning_rate=0.05,
                                        max_depth=6, subsample=0.8,
                                        colsample_bytree=0.8,
                                        random_state=42, n_jobs=-1),
    "LightGBM Regressor": lgb.LGBMRegressor(n_estimators=200, learning_rate=0.05,
                                              max_depth=6, subsample=0.8,
                                              colsample_bytree=0.8,
                                              num_leaves=63,
                                              random_state=42, n_jobs=-1,
                                              verbose=-1)
}

reg_results = {}
reg_preds   = {}

for name, model in reg_models.items():
    model.fit(X_train_r, y_train_r)
    pred = model.predict(X_test_r)
    reg_preds[name] = pred

    mse    = mean_squared_error(y_test_r, pred)
    rmse   = np.sqrt(mse)
    mae    = mean_absolute_error(y_test_r, pred)
    r2     = r2_score(y_test_r, pred)
    adj_r2 = adjusted_r2(r2, n_test, k_feat)
    cv_r2  = cross_val_score(model, X_train_r, y_train_r, cv=5, scoring="r2")

    reg_results[name] = {
        "MSE"      : mse,
        "RMSE"     : rmse,
        "MAE"      : mae,
        "R²"       : r2,
        "Adj R²"   : adj_r2,
        "CV R²"    : cv_r2.mean(),
        "CV R² std": cv_r2.std()
    }

    print(f"\n📌 {name}")
    print(f"  MSE    : {mse:.4f}")
    print(f"  RMSE   : {rmse:.4f}")
    print(f"  MAE    : {mae:.4f}")
    print(f"  R²     : {r2:.4f}")
    print(f"  Adj R² : {adj_r2:.4f}")
    print(f"  CV R²  : {cv_r2.mean():.4f} ± {cv_r2.std():.4f}")

# --- Regression Comparison Charts ---
metrics_to_plot = ["MSE", "RMSE", "MAE", "R²", "Adj R²", "CV R²"]
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()
colors = ["#E07B54", "#4C72B0"]

for i, metric in enumerate(metrics_to_plot):
    vals = [reg_results[m][metric] for m in reg_models]
    bars = axes[i].bar(reg_models.keys(), vals, color=colors)
    axes[i].set_title(metric)
    axes[i].set_xticklabels(reg_models.keys(), rotation=15, ha="right")
    for bar, val in zip(bars, vals):
        axes[i].text(bar.get_x() + bar.get_width() / 2,
                     bar.get_height() * 1.01,
                     f"{val:.3f}", ha="center", fontsize=9)

plt.suptitle("Reward Prediction — Regression Metrics Comparison", fontweight="bold", fontsize=13)
plt.tight_layout()
plt.show()

# --- Actual vs Predicted ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, (name, pred) in zip(axes, reg_preds.items()):
    ax.scatter(y_test_r, pred, alpha=0.3, s=10, color="#4C72B0")
    ax.plot([y_test_r.min(), y_test_r.max()],
            [y_test_r.min(), y_test_r.max()], "r--", lw=2)
    ax.set_xlabel("Actual Reward")
    ax.set_ylabel("Predicted Reward")
    ax.set_title(f"Actual vs Predicted\n{name}")
plt.tight_layout()
plt.show()

# --- LightGBM Feature Importance ---
lgb_feat = pd.Series(reg_models["LightGBM Regressor"].feature_importances_,
                     index=X_reg.columns)
print("\n🔍 أهم 10 Features — LightGBM Regressor:")
print(lgb_feat.sort_values(ascending=False).head(10))

plt.figure(figsize=(8, 5))
lgb_feat.sort_values(ascending=False).head(10).plot(kind="bar", color="#4C72B0")
plt.title("Top 10 Feature Importances — LightGBM Regressor")
plt.ylabel("Importance")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()


# ============================================================
# 🟣 HYBRID MODEL — CNN + MLP (Episode_Result Classification)
# ============================================================
# ⚠️ هذا القسم مستقل تماماً عن كل النماذج السابقة
# ⚠️ لا يؤثر على أي نتائج أو بيانات من الأقسام أعلاه
# ⚠️ يستخدم فقط الصفوف التي تحتوي على صورة صالحة في CNN
# ============================================================

print("\n" + "=" * 55)
print("🟣 HYBRID MODEL — CNN + MLP (Episode_Result)")
print("=" * 55)

import os
import gc
import zipfile
import tensorflow as tf
from tensorflow.keras import layers, Model, Input
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import Image as PILImage

# ──────────────────────────────────────────────
# ① فك ضغط ZIP (تلقائي — يتخطى إذا موجود مسبقاً)
# ──────────────────────────────────────────────
ZIP_PATH      = "/content/1111.zip"           # ← رفع الملف هنا في Colab
EXTRACT_DIR   = "/content/dataset_images"     # ← مجلد الاستخراج
CSV_PATH      = "/content/skycipher_dataset_20260512_125542.csv"

print("📦 فك ضغط الصور...")
if not os.path.exists(EXTRACT_DIR):
    with zipfile.ZipFile(ZIP_PATH, "r") as z:
        z.extractall("/content/")
    print(f"  ✅ تم الاستخراج إلى: /content/")
else:
    print(f"  ✅ المجلد موجود مسبقاً: {EXTRACT_DIR}")

# ──────────────────────────────────────────────
# ② تحميل CSV الجديد وربط مسارات الصور
# ──────────────────────────────────────────────
print("\n📥 تحميل CSV...")
df_sky = pd.read_csv(CSV_PATH)

# تطبيع مسار الصور: تحويل backslash → forward slash
# ثم بناء المسار الكامل داخل Colab
def normalize_img_path(raw_path):
    """تحويل المسار النسبي إلى مسار كامل داخل /content/"""
    if pd.isna(raw_path) or str(raw_path).strip() == "":
        return None
    clean = str(raw_path).replace("\\", "/").strip()
    # إزالة أي بادئة dataset_images/ مكررة
    if not clean.startswith("/content/"):
        full = os.path.join("/content", clean)
    else:
        full = clean
    return full

df_sky["full_img_path"] = df_sky["image_path"].apply(normalize_img_path)

# ──────────────────────────────────────────────
# ③ تحديد الصفوف التي تحتوي على صورة صالحة فعلاً
# ──────────────────────────────────────────────
def path_exists(p):
    return p is not None and os.path.isfile(p)

df_sky["has_image"] = df_sky["full_img_path"].apply(path_exists)

total_rows   = len(df_sky)
rows_w_image = df_sky["has_image"].sum()
print(f"  إجمالي الصفوف        : {total_rows:,}")
print(f"  صفوف لها صورة صالحة : {rows_w_image:,}")
print(f"  صفوف بدون صورة       : {total_rows - rows_w_image:,} (محذوفة من CNN فقط)")

# df_hybrid = الصفوف التي لها صورة فقط (للـ Hybrid)
df_hybrid = df_sky[df_sky["has_image"]].copy().reset_index(drop=True)

# ──────────────────────────────────────────────
# ④ Label Encoding للـ Hybrid
# ──────────────────────────────────────────────
LABEL_COL = "episode_outcome"   # عمود الهدف في CSV الجديد

df_hybrid["label"] = df_hybrid[LABEL_COL].map({
    "success": 1,
    "timeout": 0,
    "crash"  : 0,
    "stopped": 0
}).fillna(-1).astype(int)

# إزالة أي label غير معروف (-1)
df_hybrid = df_hybrid[df_hybrid["label"] >= 0].reset_index(drop=True)
print(f"\n  Label distribution:\n{df_hybrid['label'].value_counts()}")

# ──────────────────────────────────────────────
# ⑤ تحضير الـ Tabular Features للـ MLP Branch
#    نفس نهج الكود الأصلي تماماً — بدون أي تغيير في المنطق
# ──────────────────────────────────────────────
# الأعمدة المستبعدة (نفس منطق الكود الأصلي)
DROP_COLS = ["episode", "step", "reward", "episode_outcome",
             "action_x", "action_y", "action_z",
             "target_dist", "rel_x", "rel_z", "rel_y",
             "image_path", "full_img_path", "has_image", "label",
             "row_id", "frame_id"]

tab_cols = [c for c in df_hybrid.columns if c.lower() not in [d.lower() for d in DROP_COLS]]

df_tab = df_hybrid[tab_cols].copy()

# حذف أعمدة unique IDs
for col in df_tab.columns:
    if df_tab[col].nunique() == len(df_tab):
        df_tab = df_tab.drop(col, axis=1)

df_tab = pd.get_dummies(df_tab)
df_tab = df_tab.fillna(df_tab.median(numeric_only=True))

X_tab  = df_tab.values.astype(np.float32)
y_hyb  = df_hybrid["label"].values
paths  = df_hybrid["full_img_path"].values

print(f"\n  Tabular features shape : {X_tab.shape}")
print(f"  Labels shape           : {y_hyb.shape}")

# ──────────────────────────────────────────────
# ⑥ Train / Test Split للـ Hybrid (NO LEAKAGE FIX)
# ──────────────────────────────────────────────

# تأكد أن episode موجود
if "episode" not in df_hybrid.columns:
    raise ValueError("❌ episode column not found in df_hybrid!")

groups = df_hybrid["episode"].values

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

train_idx, test_idx = next(gss.split(X_tab, y_hyb, groups))

X_tab_tr = X_tab[train_idx]
X_tab_te = X_tab[test_idx]

paths_tr = paths[train_idx]
paths_te = paths[test_idx]

y_tr = y_hyb[train_idx]
y_te = y_hyb[test_idx]

print("✅ GroupShuffleSplit done using episode (NO leakage)")
print("Train samples:", len(train_idx))
print("Test samples :", len(test_idx))

# Scaling للـ tabular — fit على train فقط
scaler_tab = StandardScaler()
X_tab_tr = scaler_tab.fit_transform(X_tab_tr).astype(np.float32)
X_tab_te = scaler_tab.transform(X_tab_te).astype(np.float32)

print(f"\n  Train samples: {len(y_tr):,} | Test samples: {len(y_te):,}")

# ──────────────────────────────────────────────
# ⑦ Data Pipeline — tf.data (كفاءة عالية + RAM منخفض)
# ──────────────────────────────────────────────
IMG_SIZE   = (128, 128)   # EfficientNetB0 قادر على هذا الحجم
BATCH_SIZE = 16           # صغير لتوفير RAM
AUTOTUNE   = tf.data.AUTOTUNE

def load_image(path, img_size=IMG_SIZE):
    """قراءة صورة وتحويلها إلى tensor بحجم محدد"""
    img_raw = tf.io.read_file(path)
    img     = tf.image.decode_png(img_raw, channels=3)
    img     = tf.image.resize(img, img_size)
    img     = tf.cast(img, tf.float32) / 255.0
    return img

def make_dataset(paths, tab_feats, labels, shuffle=False):
    """بناء tf.data.Dataset يجمع صورة + tabular + label"""
    path_ds = tf.data.Dataset.from_tensor_slices(paths)
    img_ds  = path_ds.map(load_image, num_parallel_calls=AUTOTUNE)
    tab_ds  = tf.data.Dataset.from_tensor_slices(tab_feats)
    lbl_ds  = tf.data.Dataset.from_tensor_slices(labels)

    ds = tf.data.Dataset.zip(((img_ds, tab_ds), lbl_ds))
    if shuffle:
        ds = ds.shuffle(buffer_size=1000, seed=42)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(paths_tr, X_tab_tr, y_tr, shuffle=True)
test_ds  = make_dataset(paths_te, X_tab_te, y_te,  shuffle=False)

# ──────────────────────────────────────────────
# ⑧ بناء الـ Hybrid Model (CNN + MLP)
# ──────────────────────────────────────────────
n_tab_feats = X_tab_tr.shape[1]

# --- CNN Branch — EfficientNetB0 كـ Feature Extractor ---
img_input = Input(shape=(*IMG_SIZE, 3), name="image_input")
# Image Augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])

x = data_augmentation(img_input)
base_model = EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_tensor=x,
    pooling="avg"
)
base_model.trainable = False   # Freeze في البداية — Transfer Learning

cnn_out = layers.Dense(128, activation="relu",  name="cnn_dense")(base_model.output)
cnn_out = layers.Dropout(0.5, name="cnn_drop")(cnn_out)

# --- MLP Branch — Tabular Data ---
tab_input = Input(shape=(n_tab_feats,), name="tabular_input")
mlp_out   = layers.Dense(256, activation="relu",  name="mlp_dense1")(tab_input)
mlp_out   = layers.BatchNormalization(name="mlp_bn1")(mlp_out)
mlp_out   = layers.Dropout(0.4, name="mlp_drop1")(mlp_out)
mlp_out   = layers.Dense(128, activation="relu",  name="mlp_dense2")(mlp_out)
mlp_out   = layers.Dropout(0.3, name="mlp_drop2")(mlp_out)

# --- Fusion — Concatenation ---
fused = layers.Concatenate(name="fusion")([cnn_out, mlp_out])
fused = layers.Dense(128, activation="relu",  name="fused_dense1")(fused)
fused = layers.Dropout(0.5, name="fused_drop")(fused)
fused = layers.Dense(64,  activation="relu",  name="fused_dense2")(fused)

# --- Output Layer ---
output = layers.Dense(1, activation="sigmoid", name="output")(fused)

hybrid_model = Model(
    inputs=[img_input, tab_input],
    outputs=output,
    name="CNN_MLP_Hybrid"
)

hybrid_model.summary()

# ──────────────────────────────────────────────
# ⑨ Compile
# ──────────────────────────────────────────────
hybrid_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=["accuracy",
             tf.keras.metrics.AUC(name="auc")]
)

# ──────────────────────────────────────────────
# ⑩ Callbacks
# ──────────────────────────────────────────────
callbacks = [
    EarlyStopping(monitor="val_loss", patience=5,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                      patience=3, min_lr=1e-6, verbose=1)
]

# ──────────────────────────────────────────────
# ⑪ Phase 1 — تدريب الطبقات الجديدة فقط (base frozen)
# ──────────────────────────────────────────────
print("\n🏋️ Phase 1 — Training head layers (CNN frozen)...")
history_p1 = hybrid_model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=15,
    callbacks=callbacks,
    verbose=1
)

# ──────────────────────────────────────────────
# ⑫ Phase 2 — Fine-tuning: فتح آخر 20 طبقة من EfficientNet
# ──────────────────────────────────────────────
print("\n🔓 Phase 2 — Fine-tuning top layers of EfficientNetB0...")
base_model.trainable = True
fine_tune_from = len(base_model.layers) - 10

for layer in base_model.layers[:fine_tune_from]:
    layer.trainable = False

hybrid_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),   # LR أقل للـ fine-tuning
    loss="binary_crossentropy",
    metrics=["accuracy",
             tf.keras.metrics.AUC(name="auc")]
)

history_p2 = hybrid_model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=20,
    callbacks=callbacks,
    verbose=1
)

# ──────────────────────────────────────────────
# ⑬ التقييم النهائي
# ──────────────────────────────────────────────
print("\n" + "=" * 55)
print("📊 HYBRID MODEL — Final Evaluation")
print("=" * 55)

# جمع predictions
y_prob_hyb = hybrid_model.predict(test_ds, verbose=0).flatten()
y_pred_hyb = (y_prob_hyb >= 0.5).astype(int)

acc_hyb    = accuracy_score(y_te, y_pred_hyb)
f1_mac_hyb = f1_score(y_te, y_pred_hyb, average="macro")
f1_wei_hyb = f1_score(y_te, y_pred_hyb, average="weighted")
auc_hyb    = roc_auc_score(y_te, y_prob_hyb)

print(f"\n📌 CNN + MLP Hybrid")
print(f"  Accuracy      : {acc_hyb:.4f}")
print(f"  F1 (macro)    : {f1_mac_hyb:.4f}")
print(f"  F1 (weighted) : {f1_wei_hyb:.4f}")
print(f"  ROC-AUC       : {auc_hyb:.4f}")
print()
print(classification_report(y_te, y_pred_hyb,
      target_names=["Fail (0)", "Success (1)"]))

# Confusion Matrix
cm_hyb = confusion_matrix(y_te, y_pred_hyb)
plt.figure(figsize=(5, 4))
sns.heatmap(cm_hyb, annot=True, fmt="d", cmap="Purples",
            xticklabels=["Fail", "Success"],
            yticklabels=["Fail", "Success"])
plt.title("Confusion Matrix — CNN + MLP Hybrid")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.tight_layout()
plt.show()

# ──────────────────────────────────────────────
# ⑭ Training History Plot
# ──────────────────────────────────────────────
# دمج تاريخ التدريب من Phase 1 و Phase 2
acc_hist  = history_p1.history["accuracy"]      + history_p2.history["accuracy"]
val_acc   = history_p1.history["val_accuracy"]  + history_p2.history["val_accuracy"]
loss_hist = history_p1.history["loss"]          + history_p2.history["loss"]
val_loss  = history_p1.history["val_loss"]      + history_p2.history["val_loss"]
phase_split = len(history_p1.history["accuracy"])

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(acc_hist,  label="Train Accuracy",  color="#4C72B0")
axes[0].plot(val_acc,   label="Val Accuracy",    color="#E07B54")
axes[0].axvline(phase_split - 1, color="gray", linestyle="--", label="Fine-tune start")
axes[0].set_title("Hybrid Model — Accuracy")
axes[0].set_xlabel("Epoch")
axes[0].legend()

axes[1].plot(loss_hist, label="Train Loss",  color="#4C72B0")
axes[1].plot(val_loss,  label="Val Loss",    color="#E07B54")
axes[1].axvline(phase_split - 1, color="gray", linestyle="--", label="Fine-tune start")
axes[1].set_title("Hybrid Model — Loss")
axes[1].set_xlabel("Epoch")
axes[1].legend()

plt.suptitle("CNN + MLP Hybrid — Training History", fontweight="bold")
plt.tight_layout()
plt.show()

# ──────────────────────────────────────────────
# ⑮ تنظيف الذاكرة بعد الانتهاء
# ──────────────────────────────────────────────
del train_ds, test_ds
gc.collect()
tf.keras.backend.clear_session()
print("\n✅ تم تنظيف الذاكرة بعد انتهاء Hybrid Model.")
```
